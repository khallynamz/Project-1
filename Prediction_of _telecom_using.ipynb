# The dataset reveals missing values in both numerical and categorical fields, Analysis indicates that these gaps arise from the characteristics of customers who use either only phone services, only internet services,or both.
# Customers who exclusively use phone services, for example, have blank fields in columns relevant to internet services, as those attributes do not apply to them.
# A suitable approach for addressing these missing values is to fill numerical gaps with the mean of each column and the categorical gaps with "0" (indicating "Not Available").¶

```python
#fill the missing value in the data with mean,
mean_value=data['Avg Monthly GB Download'].mean()
```

```python
#Using the mean value to replace the missing records
data['Avg Monthly GB Download'].fillna(mean_value,
                                    inplace=True)
```

```python
#fill the missing value in the data with mean,and replace the missing value with the mean 
mean_value=data['Avg Monthly Long Distance Charges'].mean()
data['Avg Monthly Long Distance Charges'].fillna(mean_value,  inplace=True)
```

```python
# function to create daignostic plots
def plot_boxplot (df,ft):
    df.boxplot(column=[ft])
    plt.grid(False)
        
    # hist
    plt.subplot(1,2,1)
    plt.hist(df[ft], alpha=0.2)
    
    # boxplot
    plt.subplot(1,2,2)
    sns.boxplot(df[ft])
    plt.show()
```

```python
# lets find IQR(inter quantile range)

cat = df.select_dtypes(include='object').columns #object type columns
num = list(df.select_dtypes(exclude='object').columns) #numerical type columns


for ft in num:
    Q1=df[ft].quantile(0.25)
    Q3=df[ft].quantile(0.75)
    IQR= Q3-Q1
    lower_bound=Q1-(1.5*IQR)
    upper_bound=Q3+(1.5*IQR)
    
    # store index of variables
    df[ft]=df[ft].mask(df[ft]<lower_bound,lower_bound, )
    df[ft]=df[ft].mask(df[ft]>upper_bound,upper_bound, )
```

# Apply Survival analysis to predict Churn_Timining

```python
from lifelines import KaplanMeierFitter

# Fit Kaplan-Meier Model
kmf = KaplanMeierFitter()
kmf.fit(durations=X1['Tenure in Months'], event_observed=y1)

# Find time when survival probability drops below a threshold
threshold = 0.5  # Example: 50% churned
churn_time = kmf.survival_function_.loc[kmf.survival_function_['KM_estimate'] <= threshold].index[0]

print(f"Estimated time to churn for 50% of customers: {churn_time} months")

# Plot Survival Function
kmf.plot_survival_function(title="Kaplan-Meier Survival Curve")
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.datasets import make_classification
# Step 1: Generate initial synthetic data
X_train_res, y_train_res = make_classification(n_samples=1000,  # Training data size
                                               n_features=X_train_res.shape[1],   # Number of features
                                               n_informative=5,
                                               n_classes=2,     # Binary classification
                                               random_state=42)

# Convert to DataFrame for consistent structure
X_train_res = pd.DataFrame(X_train_res, columns=[f"Feature_{i}" for i in range(28)])

# Step 2: Initialize SGDClassifier for online learning
sgd_clf = SGDClassifier(loss='log_loss',  # Logistic regression loss
                        max_iter=1,       # Single iteration per partial_fit
                        warm_start=True,  # Allows incremental learning
                        random_state=42)

# Step 3: Perform initial incremental training
sgd_clf.partial_fit(X_train_res, y_train_res, classes=[0, 1])
print("Initial training complete!")

# Step 4: Generate new streaming data with the same feature structure
X_new, y_new = make_classification(n_samples=100,     # Simulate 100 new samples
                                   n_features=X_train_res.shape[1],   # Must match training feature count
                                   n_informative=5,
                                   n_classes=2,     # Match number of classes
                                   random_state=42)

# Convert new data into DataFrame
X_new = pd.DataFrame(X_new, columns=X_train_res.columns)
print("New batch shape:", X_new.shape)

# Step 5: Incremental retraining with new data
sgd_clf.partial_fit(X_new, y_new)
print("Incremental retraining complete!")

# Step 6: Test the model on new data (optional)
predictions = sgd_clf.predict(X_new)
print("Predictions on new data:", predictions)
```
